```python
x_in = keras.Input(shape=(seize), batch_size = 1)
# x = layers.GlobalMaxPooling2D()(x_in)
x = layers.Flatten(input_shape=(seize,))(x_in)
x = layers.Dense(40, kernel_regularizer='l2', activation="relu")(x)
x = layers.Dropout(0.5)(x)
x = layers.Dense(40, kernel_regularizer='l2', activation="relu")(x)
x = layers.Dropout(0.5)(x)
# x = layers.Dense(330, activation="sigmoid")(x)
# # x = layers.Dropout(0.7)(x)
# x = layers.Dense(920, activation="sigmoid")(x)
# # x = layers.Dropout(0.7)(x)
# x = layers.Dense(920, activation="sigmoid")(x)
# # x = layers.Dropout(0.7)(x)
x_out = layers.Dense(1, kernel_regularizer='l2', activation="sigmoid")(x)
# x_out = layers.Softmax()(x)


model = keras.Model(inputs=x_in, outputs=x_out)
```

```python


 model.compile(loss='BinaryCrossentropy', optimizer=keras.optimizers.Adam(learning_rate=1e-5), metrics='Accuracy')
history = model.fit(X_train, Y_train, batch_size=1, epochs=30, validation_split = 0.05)
```
